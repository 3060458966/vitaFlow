### Loss

- For general image classification we prefer "pixel-wise cross entropy loss", which ensure that all the pixel are learned equally 
- In case image segmentation, one masked segmentation can overwhelmly bias the other one based on the size and shae of the objects under concentration
- To avoid this imbalance in learning, a loss called "[Dice Cofficient is used](https://en.wikipedia.org/wiki/S%C3%B8rensen%E2%80%93Dice_coefficient)" is used
- Second loss is IoU : Intersection of Union Loss
