```python
import os
import csv
import pickle
import pandas as pd

from glob import glob
import sklearn_crfsuite

full_path = '/Users/sampathm/devbox/vitaFlow/vitaflow/standoff2conll/receipt_mock_data'
```

```python
def strip_iob(iob_tag):
    tag = iob_tag.replace("B-", "")
    tag = tag.replace("I-", "")
    return tag


def is_new_tag(prev, current):
    if "O" in prev:
        prev_t, prev_w = "O", "O"
    else:
        prev_t, prev_w = prev.split("-")
    if "O" in current:
        current_t, current_w = "O", "O"
    else:
        current_t, current_w = current.split("-")

    if prev_w != current_w:
        return True
    else:
        if prev_t == "B" and current_t == "I":
            return False
        elif prev_t == "I" and current_t == "B":
            return True
        elif prev_t == "I" and current_t == "I":
            return False
        else:
            return False

def word2features(sent, i):
    word = sent[i][0]
    features = {
        'bias': 1.0,
        'word.lower()': word.lower() if type(word) is str else False,
        'word[-3:]': word[-3:] if type(word) is str else word,
        'word[-2:]': word[-2:] if type(word) is str else word,
        'word.isupper()': word.isupper() if type(word) is str else False,
        'word.istitle()': word.istitle() if type(word) is str else False,
        'word.isdigit()': word.isdigit() if type(word) is str else False
    }
    if i > 0:
        word1 = sent[i - 1][0]
        features.update({
            '-1:word.lower()': word1.lower() if type(word1) is str else False,
            '-1:word.istitle()': word1.istitle() if type(word1) is str else False,
            '-1:word.isupper()': word1.isupper() if type(word1) is str else False
        })
    else:
        features['BOS'] = True

    if i < len(sent) - 1:
        word1 = sent[i + 1][0]
        features.update({
            '+1:word.lower()': word1.lower() if type(word1) is str else False,
            '+1:word.istitle()': word1.istitle() if type(word1) is str else False,
            '+1:word.isupper()': word1.isupper() if type(word1) is str else False
        })
    else:
        features['EOS'] = True
    return features


def sent2features(sent):
    return [word2features(sent, i) for i in range(len(sent))]


def sent2labels(sent):
    return [label for token, label in sent]


def sent2tokens(sent):
    return [token for token, label in sent]


def _get_IOB_data(file):
    df = pd.read_csv(file, sep="\t", quoting=csv.QUOTE_NONE)
    data = list(zip(*[df[c].values.tolist() for c in ['0', '1']]))
    return data
    

def get_IOB_data(path):
    collected_data = []
    fname = path + '/*csv'
    print(f'Reading {fname}')
    for file in glob(fname):
        data = _get_IOB_data(file)
        collected_data.append(data)
        # print('Reading {}'.format(file))
    return collected_data


def training_crf():
    print('Training Data,..')
    train_sent = get_IOB_data(full_path + "/train_csv")
    X_train = [sent2features(s) for s in train_sent]
    y_train = [sent2labels(s) for s in train_sent]
    print('Training Data,..  Loaded!!')
    crf = sklearn_crfsuite.CRF(
        algorithm='lbfgs',
        c1=0.06,
        c2=0.3,
        max_iterations=100,
        all_possible_transitions=True
    )
    print('Training Score.. ')
    crf.fit(X_train, y_train)
    pickle.dump(crf, open('crf_model.pickle', 'wb'))
    print('Training Score.. {}'.format(crf.score(X_train, y_train)))

def testing_crf():
    print('Testing Data,..')
    test_sent  = get_IOB_data(full_path + "/test_csv")
    X_test = [sent2features(s) for s in test_sent]
    y_test = [sent2labels(s) for s in test_sent]
    print('Testing Data,.. Loaded')
    crf = pickle.load(open('crf_model.pickle', 'rb'))
    print('Training Score.. ')
    print('Training Score.. {}'.format(crf.score(X_test, y_test)))
    
def predict_crf():
    print('Val Data,..')
    val_sent  = get_IOB_data(full_path + "/val_csv")
    X_test = [sent2features(s) for s in test_sent]
    y_test = [sent2labels(s) for s in test_sent]
    print('Val Data,.. Loaded')
    crf = pickle.load(open('crf_model.pickle', 'rb'))
    print('Val Score.. ')
    print('Val Score.. {}'.format(crf.score(X_test, y_test)))
```
